```python
import numpy as np
x=np.arange(24).reshape(4,6)
x
```

```python
print(x[1][2], x[1,2])
```

```python
x[[1,2],[2,3]] #1행에 2열, 2행에 3열
```

```python
x[[1,2]] #1행과 2행
```

```python
X_train[...,np.newaxis].shape
```

newaxis는 내부적으로 reshape보다 연산이 빠름(아주 미세한 차이긴 함), 또한 reshape은 (4,1)을 (2,2)로 변환할 순 있어도 (3,3)으로 변환할 순 없음, 재배열하는 느낌

```python
a=np.array([1, 2, 3, 4, 5])
b=np.array([5, 4, 3])

#a+b #오류남

a=a[...,np.newaxis] #a가 (5,1)꼴이 되면서 계산이 가능해짐, a+b는 (5,3)꼴
a+b
```

## ML for image

```python
from sklearn.datasets import load_digits
from sklearn.neighbors import KNeighborsClassifier

#데이터 불러오기
digits=load_digits()

#모델 생성 및 학습
knn=KNeighborsClassifier()
knn.fit(digits.data, digits.target)

#모델 예측
knn.predict(digits.data[0,None])
```

```python
digits.data[0].shape
```

```python
digits.data[0].reshape(8,8) #1x64의 1차원 데이터를 8x8의 2차원으로 만듦
```

```python
plt.imshow(digits.data[0].reshape(8,8))
```

왜 흑백이미지를 컬러로 보여줄까? 단순 숫자 패턴인 array만으론 color space를 모르기 때문에 matplotlib의 기본적인 color map을 입혀준 것, 컴퓨터는 흑백이미지인지/rgb인지/hsv인지 등 모름

```python
plt.imshow(digits.data[0].reshape(8,8), cmap='gray')
```

```python
plt.imshow(digits.data[0,None], cmap='gray')
```

인접한 픽셀이 같이 있어야 의미가 있는 거지, 1차원으로 변환하면(=일렬로 펴주면) 그 의미를 잃게 됨, 로컬정보를 갖기 때문임

그러나 딥러닝에서 Dense layer는 1차원만 받기 때문에 1차원 데이터로 만들어주는 과정을 거치게 됨, 그것이 바로 Flatten

```python
import tensorflow as tf
x=tf.keras.layers.Flatten()
data=digits.data.reshape(1797,8,8)
x(data)
```
